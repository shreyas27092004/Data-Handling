```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x));
ax.plot(x,np.cos(x));
ax.plot(x,np.tan(x));
ax.set_title('simple plot')
ax.set_ylabel('Amplitude')
ax.set_xlabel('Wavelength')
```

numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)[source]
Return evenly spaced numbers over a specified interval.

Returns num evenly spaced samples, calculated over the interval [start, stop].

The endpoint of the interval can optionally be excluded.

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x), label = 'sin', color = 'purple');
ax.plot(x,np.cos(x), label = 'cos', color = 'green');
#ax.plot(x,np.tan(x));
ax.set_title('Multiple Lines')
ax.set_ylabel('Amplitude')
ax.set_xlabel('Wavelength')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
print(x)
ax.plot(x,np.sin(x), label = 'sin', linestyle = 'dashed');
ax.plot(x,np.cos(x), label = 'cos', linestyle = 'dotted');
ax.plot(x,np.sin(x+1), label = 'sin+1', linestyle = 'dashdot');
#ax.plot(x,np.tan(x));
ax.set_title('Multiple Lines')
ax.set_ylabel('Amplitude')
ax.set_xlabel('Wavelength')
ax.legend()
plt.show()
```
