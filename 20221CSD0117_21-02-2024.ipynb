```python
df=pd.read_csv("/BL-Flickr-Images-Book (1).csv")
df.tail()
```

```python
to_drop=['Edition Statement','Shelfmarks']
df.drop(to_drop,inplace=True,axis=1)
df.head()
```

```python
df.set_index('Identifier',inplace=True)
df.head()
```

```python
unwanted_characters=['[',',','-']
def clean_dates(item):
  dop=str(item.loc['Date of Publication'])
  if dop=='nan' or dop[0]=='[':
    return np.NAN
  for charcater in unwanted_characters:
    if charcater in dop:
      character_index=dop.find(charcater)
      dop=dop[:character_index]
  return dop
df['Date of Publication']=df.apply(clean_dates,axis=1)
```
