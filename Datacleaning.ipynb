```python
# @title Date of Publication vs count()

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  from matplotlib import pyplot as plt
  import seaborn as sns
  palette = list(sns.palettes.mpl_palette('Dark2'))
  counted = (series['Date of Publication']
                .value_counts()
              .reset_index(name='counts')
              .rename({'index': 'Date of Publication'}, axis=1)
              .sort_values('Date of Publication', ascending=True))
  xs = counted['Date of Publication']
  ys = counted['counts']
  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Date of Publication', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date of Publication')
_ = plt.ylabel('count()')
```

```python
unwanted_characters = ['[', ',', '-']
def clean_dates(item):
  dop= str(item.loc['Date of Publication'])

  if dop =='nan' or dop[0] == '[':
    return np.NaN

  for character in unwanted_characters:
    if character in dop:
      character_index=dop.find(character)
      dop = dop[:character_index]
  return dop
df['Date of Publication']=df.apply(clean_dates, axis = 1)
```
